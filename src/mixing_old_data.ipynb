```python
import pandas as pd
```

```python
old = pd.read_csv("../data/2017.csv")
```

```python
df = pd.read_csv("../data/cleaned_flights_with_weather.csv")
```

```python
df["fl_date"] = pd.to_datetime(df["fl_date"])
df.sort_values(by="fl_date")
```

```python
df["last_year"] = df["fl_date"] - pd.DateOffset(years=1)
```

```python
old.columns = [s.lower() for s in old.columns]
```

```python
old.head()
```

```python
old.columns
```

```python
o=old[["op_carrier", "arr_delay", "weather_delay", "nas_delay", "security_delay", "late_aircraft_delay", "carrier_delay"]].groupby("op_carrier").mean().reset_index().rename(columns={"arr_delay": "mean_arr_delay", "weather_delay": "mean_weather_delay", "nas_delay": "mean_nas_delay", "security_delay": "mean_security_delay", "late_aircraft_delay": "mean_late_aircraft_delay", "carrier_delay": "mean_carrier_delay"})
```

```python
std = old[["op_carrier", "arr_delay", "weather_delay", "nas_delay", "security_delay", "late_aircraft_delay", "carrier_delay"]].groupby("op_carrier").std().reset_index().rename(columns={"arr_delay": "std_arr_delay", "weather_delay": "std_weather_delay", "nas_delay": "std_nas_delay", "security_delay": "std_security_delay", "late_aircraft_delay": "std_late_aircraft_delay", "carrier_delay": "std_carrier_delay"}).drop("op_carrier", axis=1)
```

```python
o = pd.concat((o, std), axis=1)
```

```python
df.shape
```

```python
d=df.merge(o, left_on=["mkt_carrier"], right_on=["op_carrier"]).drop(columns=["op_carrier"])
```

```python
d
```

```python
d.to_csv("../data/cleaned_with_mean_arr_delay.csv", index=False)
```

```python
old["fl_date"] = pd.to_datetime(old["fl_date"])
```

```python
s = df.merge(right=df[["fl_date", "arr_delay", "origin", "dest", "mkt_unique_carrier"]], how="inner", left_on=["last_year", "origin", "dest", "mkt_unique_carrier"], right_on=["fl_date", "origin", "dest", "mkt_unique_carrier"], suffixes=["", "_drop"])
```

```python
s = s.rename(columns={"arr_delay_drop": "arr_delay_last_year"})
```

```python
s.drop([col for col in s.columns if "drop" in col], axis=1, inplace=True)
```

```python
m = df.merge(old[["fl_date", "origin", "dest", "op_carrier", "arr_delay"]], how="inner", left_on=["last_year", "origin","dest", "op_unique_carrier"], right_on=["fl_date", "origin", "dest","op_carrier"], suffixes=["", "_drop"])
```

```python
m.shape
```

```python
m = m.rename(columns={"arr_delay_drop": "arr_delay_last_year"})
```

```python
m = m.drop([col for col in m.columns if "drop" in col], axis=1)
```

```python
old = m.drop_duplicates(subset=["fl_date", "origin", "dest", "op_unique_carrier"])
```

```python
old = old.drop(columns=["op_carrier", "dup", "last_year"])
```

```python
old = old.reset_index(drop=True)
```

```python
old.info()
```

```python
old = old.fillna(0)
```

```python
old.info()
```

```python
old.to_csv("../data/flights_with_previous_year_data.csv", index=False)
```
