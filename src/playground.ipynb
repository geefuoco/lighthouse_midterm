```python
df_flights = get_data_sample(table_name="flights")
```

```python
df_flights.columns
```

```python
means = df_flights[["mkt_carrier", "arr_delay", "nas_delay", "late_aircraft_delay"]].groupby("mkt_carrier").mean().rename(columns={"arr_delay": "mean_arr_delay", "nas_delay": "mean_nas_delay", "late_aircraft_delay": "mean_late_aircraft_delay"})
```

```python
m = df.merge(means, on=["mkt_carrier"], suffixes=["", "_drop"])
```

```python
m.info()
```

```python
m = m.sort_values(by=["fl_date"]).reset_index(drop=True)
```

```python
import pandas as pd

w = pd.read_csv("../data/weather/WeatherEvents_Jan2016-Dec2021.csv")
```

```python
w.columns
```

```python
w["fl_date"] = pd.to_datetime(w["StartTime(UTC)"]).dt.date
```

```python
w = w.loc[(w["fl_date"] >= pd.to_datetime("2020-01-01")) & (w["fl_date"] < pd.to_datetime("2020-01-08"))]
```

```python
w["origin_city_name"] = w["City"] + ", " + w["State"]
```

```python
m["fl_date"] = pd.to_datetime(m["fl_date"])
```

```python
w["fl_date"] = pd.to_datetime(w["fl_date"])
```

```python
wm = m.merge(w[["fl_date", "origin_city_name", "Precipitation(in)"]], on=["fl_date", "origin_city_name"])
```

```python
wm = wm.drop_duplicates()
```
