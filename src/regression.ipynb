### Splitting Data

```python
cols = ["arr_delay", "cancelled", "carrier_delay", "weather_delay", "nas_delay", "security_delay", "late_aircraft_delay"]
# X = df.drop(columns=cols)
X = df[related_columns]
y = df["arr_delay"]
```

### Feature Extraction / Selection

```python
from modules.transformation import standard_scale, hot_encode, log_transform, ordinal_encode
```

```python
X = X.drop(columns=["mkt_carrier_fl_num", "op_carrier_fl_num", "flights", "mkt_unique_carrier", "branded_code_share", "origin_city_name", "dest_city_name"])
```

```python
X["trip_duration"] = "short"
X.loc[X["distance"] > X["distance"].mean(), "trip_duration"] = "long"
```

```python
X.loc[:, ["distance","trip_duration"]]
```

```python
times = (X[["crs_dep_time", "crs_arr_time"]] / 100).astype(int)
```

```python
bin_time = [0, 6, 12, 18, 23]
catg_time = ["Night", "Morning", "Afternoon", "Evening"]
times["crs_dep_time"] = pd.cut(times["crs_dep_time"], bin_time, labels=catg_time)
times["crs_arr_time"] = pd.cut(times["crs_arr_time"], bin_time, labels=catg_time)
times = times.rename(columns={"crs_dep_time": "dep_time_of_day", "crs_arr_time": "arr_time_of_day"})
times
```

```python
X = pd.concat((X, times), axis=1)
```

#### Ordinal Encoding

```python
ordinal_encode(X, ["trip_duration"], [["short", "long"]])
```

#### Log Transform

```python
log_transform(X, cols=["crs_elapsed_time", "distance"])
```

#### Dates

```python
X["fl_date"] = pd.to_datetime(X["fl_date"])
# X["month"] = X["fl_date"].dt.month
# X["day"] = X["fl_date"].dt.day
# X["year"] = X["fl_date"].dt.year
X["day_of_week"] = X["fl_date"].dt.dayofweek
```

```python
X["winter"] = 0
X["summer"] = 0
X["autumn"] = 0
```

```python
X.loc[(X["fl_date"] >= "2017-12-21") & (X["fl_date"] < "2018-03-21"), "winter"] = 1
X.loc[(X["fl_date"] >= "2018-06-21") & (X["fl_date"] < "2018-09-21"), "summer"] = 1
X.loc[(X["fl_date"] >= "2018-09-21") & (X["fl_date"] < "2018-12-21"), "autumn"] = 1

X.loc[(X["fl_date"] >= "2018-12-21") & (X["fl_date"] < "2019-03-21"), "winter"] = 1
X.loc[(X["fl_date"] >= "2019-06-21") & (X["fl_date"] < "2019-09-21"), "summer"] = 1
X.loc[(X["fl_date"] >= "2019-09-21") & (X["fl_date"] < "2019-12-21"), "autumn"] = 1
```

### Train and Split Data

```python
from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(transformed, y, test_size=0.2, random_state=10)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

### Model

```python
from sklearn.metrics import mean_squared_error

def print_scores(X_test, y_test, model):
  y_pred = model.predict(X_test)
  print("R2 Score: {:.5f}".format(model.score(X_test, y_test)))
  print("Mean Squared Error: {:.5f}".format(mean_squared_error(y_test, y_pred)))
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(X_train, y_train)

print_scores(X_test, y_test, reg)
```

Can rule out overfitting since training data also has a poor score

```python
print_scores(X_train, y_train, reg)
```

#### Ridge Regression

```python
from sklearn.linear_model import Ridge

r = Ridge()

r.fit(X_train, y_train)

print_scores(X_test, y_test, r)
```

#### Lasso Regression

```python
from sklearn.linear_model import Lasso

l = Lasso()

l.fit(X_train, y_train)

print_scores(X_test, y_test, l)
```

#### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

rfc = RandomForestRegressor(n_jobs=-1, random_state=10, max_depth=30)

rfc.fit(X_train, y_train)

print_scores(X_test, y_test, rfc)
```

#### KNearest Neighbours

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_jobs=-1)

knn.fit(X_train, y_train)

print_scores(X_test, y_test, knn)
```
