```python
idx = df[df["mean_nas_delay"] > 31].index
df = df.drop(index=idx)
```

```python
df = df.reset_index(drop=True)
```

```python
df["fl_date"] = pd.to_datetime(df["fl_date"])
df["day_of_week"] = df["fl_date"].dt.dayofweek
```

```python
df
```

```python
X = df.drop(columns=["fl_date", "origin", "dest", "mkt_carrier"])
```

```python
from modules.transformation import standard_scale

standard_scale(X, cols=X.columns)
```

```python
from modules.models import model_loader
```

```python
gbr = model_loader.load_model("../data/models/gradient_boosting_regression")
```

```python
y_pred = gbr.predict(X)
```

```python
s = pd.Series(y_pred, name="predicted_delay")
```

```python
sample_submission = pd.concat((df, s), axis=1)
```

```python
s_cols = ["fl_date", "mkt_carrier", "mkt_carrier_fl_num", "origin", "dest", "predicted_delay"]
sample_submission = sample_submission[s_cols]
```
